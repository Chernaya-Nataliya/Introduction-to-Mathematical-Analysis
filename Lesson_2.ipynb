### –ó–∞–¥–∞—á–∞ 2

–ò—Å—Å–ª–µ–¥—É–µ–º –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –Ω–∞ –º–æ–Ω–æ—Ç–æ–Ω–Ω–æ—Å—Ç—å, –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–Ω–æ—Å—Ç—å –∏ –Ω–∞–π–¥–µ–º –ø—è—Ç—ã–π —á–ª–µ–Ω.

### –ó–∞–¥–∞—á–∞ 3
*–ù–∞ —è–∑—ã–∫–µ Python –ø—Ä–µ–¥–ª–æ–∂–∏—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º –≤—ã—á–∏—Å–ª—è—é—â–∏–π —á–∏—Å–ª–µ–Ω–Ω–æ –ø—Ä–µ–¥–µ–ª —Å —Ç–æ—á–Ω–æ—Å—Ç—å—é
$\epsilon = 10^7$
$\lim\limits_{n \to +\infty} \frac{n}{\sqrt[n]{n!}}$

```python
# –õ–æ–≥–∞—Ä–∏—Ñ–º —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª–∞: –î–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª–∞ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –ª–æ–≥–∞—Ä–∏—Ñ–º —Å—É–º–º—ã, —á—Ç–æ–±—ã –∏–∑–±–µ–∂–∞—Ç—å –ø–µ—Ä–µ–ø–æ–ª–Ω–µ–Ω–∏—è.
# –ü–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å: –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–∞–∑–Ω–∏—Ü—É –º–µ–∂–¥—É —Ç–µ–∫—É—â–∏–º –∏ –ø—Ä–µ–¥—ã–¥—É—â–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ–º, —á—Ç–æ–±—ã –¥–æ—Å—Ç–∏–≥–Ω—É—Ç—å –∑–∞–¥–∞–Ω–Ω–æ–π —Ç–æ—á–Ω–æ—Å—Ç–∏ $\epsilon$.

import math

def calculate_limit(epsilon=1e-7):
    prev_value = 0
    n = 1
    log_fact = 0  # –õ–æ–≥–∞—Ä–∏—Ñ–º —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª–∞
    while True:
        log_fact += math.log(n)
        n_fact_root = math.exp(log_fact / n)
        current_value = n / n_fact_root
        if abs(current_value - prev_value) < epsilon:
            break
        prev_value = current_value
        n += 1
    return current_value, n

limit, n = calculate_limit()
print(f"–ü—Ä–µ–¥–µ–ª –ø—Ä–∏–±–ª–∏–∑–∏—Ç–µ–ª—å–Ω–æ —Ä–∞–≤–µ–Ω {limit}, —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω—ã–π –ø—Ä–∏ n = {n}")
```

### –ó–∞–¥–∞—á–∞ 4
*–ü—Ä–µ–¥–ª–æ–∂–∏—Ç—å –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏—é –∞–ª–≥–æ—Ä–∏—Ç–º–∞, –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ –≤ –∑–∞–¥–∞–Ω–∏–∏ 3, —É—Å–∫–æ—Ä—è—é—â—É—é –µ–≥–æ —Å—Ö–æ–¥–∏–º–æ—Å—Ç—å

–î–ª—è –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏ –∞–ª–≥–æ—Ä–∏—Ç–º–∞, —É—Å–∫–æ—Ä—è—é—â–µ–≥–æ –µ–≥–æ —Å—Ö–æ–¥–∏–º–æ—Å—Ç—å, –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –º–µ—Ç–æ–¥ –°—Ç–µ—Ä–ª–∏–Ω–≥–∞ –¥–ª—è –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏–∏ —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª–∞:

$n! \approx \sqrt{2\pi n} \left(\frac{n}{e}\right)^n$

```python
import mpmath

def calculate_limit_optimized(epsilon=1e-7):
    mpmath.mp.dps = 50  # –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π –¥–ª—è —Ä–∞—Å—á–µ—Ç–æ–≤. –ó–Ω–∞—á–µ–Ω–∏–µ –≤ 50 –∑–Ω–∞–∫–æ–≤ –≤—ã–±—Ä–∞–Ω–æ –∫–∞–∫ "—Å –∑–∞–ø–∞—Å–æ–º", —á—Ç–æ–±—ã —ç—Ç–æ –Ω–µ –ø–æ–≤–ª–∏—è–ª–æ –Ω–∞ —Ç–æ—á–Ω–æ—Å—Ç—å –∏—Ç–æ–≥–æ–≤—ã—Ö —Ä–∞—Å—á–µ—Ç–æ–≤.
    prev_value = mpmath.mpf('0')
    n = 1
    while True:
        n_fact_approx = mpmath.sqrt(2 * mpmath.pi * n) * (n / mpmath.e)**n
        n_fact_root = n_fact_approx**(1/n)
        current_value = n / n_fact_root
        if abs(current_value - prev_value) < epsilon:
            break
        prev_value = current_value
        n += 1
    return current_value, n

limit, n = calculate_limit_optimized()
print(f"–û–ø—Ç–∏–º–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø—Ä–µ–¥–µ–ª –ø—Ä–∏–±–ª–∏–∑–∏—Ç–µ–ª—å–Ω–æ —Ä–∞–≤–µ–Ω {limit}, —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω—ã–π –ø—Ä–∏ n = {n}")
```
